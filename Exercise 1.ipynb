```python
#import plotting library
import matplotlib.pyplot as plt
```

```python
#create year and GDP lists
Year = list()
for x in range (1930, 2020, 10):
    Year.append(x)
    
GDP = [1.015, 1.33, 2.29, 3.266, 4.951, 6.759, 9.366, 13.131, 15.599]
```

```python
#create plot & plot labels
plt.scatter(Year, GDP)
plt.title('Real US GDP (in trillions)')
plt.xlabel('Year')
plt.ylabel('Trillions of $')

#show plot
plt.show
```
