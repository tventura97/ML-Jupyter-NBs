```python
#import lib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Question 1
#create year and GDP lists
Year = list()
for x in range (1930, 2020, 10):
    Year.append(x)
    
GDP = [1.015, 1.33, 2.29, 3.266, 4.951, 6.759, 9.366, 13.131, 15.599]
```

```python
#create plot & plot labels
plt.plot(Year, GDP, '--o', color = 'r')
plt.title('Real US GDP (in trillions)')
plt.xlabel('Year')
plt.ylabel('Trillions of $')

#show plot
plt.show
```

```python
#Question 2
#If you do a simple linear regression, you can create a simple mathematical model for the relationship between the year and US GDP.
#Create numpy arrays of the data
yrs = np.array([1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010])
yrs = yrs.reshape(yrs.size,1)
gdp = np.array([1.015, 1.33, 2.29, 3.266, 4.951, 6.759, 9.366, 13.131, 15.599])
gdp = gdp.reshape(gdp.size,1)
model = LinearRegression()
model.fit(yrs,gdp)

# generate model from the data
yrs_model = np.linspace(1930, 2010, 100)
gdp_model = model.predict(yrs_model[:, np.newaxis])

#a linear regression creates an equation in the form y = mx + b;
#we can find the slope using any two points from the generated model and then use any single point to find b.

m = (gdp_model[1] - gdp_model[0])/(yrs_model[1]-yrs_model[0])
#Find b using the very first point
b = gdp_model[0] - m*yrs_model[0]

#So, the model is
print("y = ",m,"x + ", b)

#Question 3
#plot model
plt.plot(yrs, gdp, '--o', color = 'r')
plt.title('Real US GDP (in trillions)')
plt.xlabel('Year')
plt.ylabel('Trillions of $')
plt.plot(yrs_model, gdp_model)
plt.show()
```
