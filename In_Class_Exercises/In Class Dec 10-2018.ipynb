```python
import sklearn 
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
from sklearn import datasets

#Load dataset
x,y = datasets.load_boston(return_X_y=True)
#x is data, y is MEDV (the target we want to predict)
l = len(x)
#need to augment x with one
ones = np.ones(506)
x = x.transpose()
x = np.vstack((x,ones))
x = x.transpose()

print(x.shape)
print(y.shape)
```

```python
#Note  y = xTw + b
#y = xTw (b packaged in x)
#want to find w
#b is some intercept

#According to the notes - this is the LS solution
def LSRegression(x, y):
    return inv((x.transpose() @ x))@ x.transpose() @ y
```

```python
#Try LS Regression
w = LSRegression(x,y)
y_hat = x @ w

error = 0

for i in range(len(y_hat)):
    error += (y[i] - y_hat[i])**2/l
```

```python
#Cross validation
x_split = np.array_split(x, 5)
y_split = np.array_split(y, 5)
_w = []

for i in range(0,5):
    test = x_split[i]
    train_labels = np.hstack([y_split[j] for j in range(0,5) if j != i])
    temp = [x_split[j] for j in range(0,5) if j != i]
    train = np.vstack(temp)
    _w.append(LSRegression(train, train_labels))    

errs = []
for w in _w:
    y_hat = x@w
    err = 0
    for i in range(len(y_hat)):
        err += abs(y[i] - y_hat[i])**2
    errs.append(err/506)

    
print(np.mean(errs))
    
#_w.append(LSRegression(train, y))
```
